```python
# 数据处理库
import pandas as pd
import numpy as np

# 数据预处理库
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler

```

```python
# 创建示例数据
data = {
    "Age": [23, 25, 29, np.nan, 35, 40, 120],
    "Income": [3000, 3500, 4000, 5000, np.nan, 8000, 10000],
    "Gender": ["M", "F", "F", "M", "F", "M", "M"]
}

df = pd.DataFrame(data)

print("原始数据：")
print(df)
```

```python
print("数据结构：")
print(df.info())
```

```python
print("缺失值统计：")
print(df.isnull().sum())
```

```python
# 缺失值处理： 删除

df_drop = df.dropna()

print("删除缺失值后的数据：")

print(df_drop)
```

```python
#缺失值处理：均值插值


df["Age"].fillna(df["Age"].mean(), inplace=True)
df["Income"].fillna(df["Income"].mean(), inplace=True)

print("均值填补后的数据：")
print(df)
```

```python
# 使用Z-score检测异常值
z_scores = (df["Age"] - df["Age"].mean()) / df["Age"].std()

print("Z-score:")
print(z_scores)
```

```python
# 类别变量编码 ~ one hot encoding
df_encoded = pd.get_dummies(df, columns=["Gender"])

print("编码后的数据：")
print(df_encoded)
```

```python
scaler = MinMaxScaler()

scaled_data = scaler.fit_transform(df_encoded)

df_scaled = pd.DataFrame(scaled_data, columns=df_encoded.columns)

print("归一化后的数据：")
print(df_scaled)
```

```python
scaler = StandardScaler()

standard_data = scaler.fit_transform(df_encoded)

df_standard = pd.DataFrame(standard_data, columns=df_encoded.columns)

print("标准化后的数据：")
print(df_standard)
```

```python
# 余弦相似
import numpy as np

def cosine_similarity(a, b):
    a = np.array(a)
    b = np.array(b)

    dot_product = np.dot(a, b)
    norm_a = np.linalg.norm(a)
    norm_b = np.linalg.norm(b)

    similarity = dot_product / (norm_a * norm_b)
    return similarity


# example
A = [1, 2, 3]
B = [4, 5, 6]

print(cosine_similarity(A, B))
```

```python
# 余弦相似

from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

A = np.array([[1, 2, 3]])
B = np.array([[4, 5, 6]])

sim = cosine_similarity(A, B)

print(sim)
```

```python
#文本检测相似度

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

documents = [
    "I love machine learning",
    "I like machine learning",
    "Dogs are cute"
]

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(documents)

print(X)

sim = cosine_similarity(X)

print(sim)
```

```python
# 信息熵计算

import numpy as np

def entropy(probabilities):
    probabilities = np.array(probabilities)
    return -np.sum(probabilities * np.log2(probabilities))

# example： 四个面的骰子
p = [0.25, 0.25, 0.25, 0.25]

print(entropy(p))
```

```python
# 信息熵计算
from scipy.stats import entropy

p = [0.25, 0.25, 0.25, 0.25]

print(entropy(p, base=2))
```
